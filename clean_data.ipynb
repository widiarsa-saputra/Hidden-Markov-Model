```python
def data_frame_split(df):
    array_sequens_word = []
    array_sequens_entity = []

    array_word = []
    array_entity = []
    for index, val in df.iterrows():
        if index != 0 and val['SentenceID'] != df['SentenceID'][index-1]:
            array_sequens_word.append(array_word)
            array_sequens_entity.append(array_entity)

            array_word = []
            array_entity = []
        array_word.append(str(val['Word']))
        array_entity.append(str(val['Character Named Entity Tagset']))
    return array_sequens_word, array_sequens_entity
```

```python
sequens_word, sequens_entity = data_frame_split(dataset_frame)
```

```python
save_dataframe = pd.DataFrame(list(zip(sequens_word, sequens_entity)), columns=["sequens_word", "sequens_entity"])
```

```python
def split_conllu_data(conllu_file):
    words = []
    pos_tags = []
    
    # Membuka file dengan encoding UTF-8
    with open(conllu_file, 'r', encoding='utf-8') as f:
        sentence_words = []
        sentence_pos_tags = []
        
        for line in f:
            line = line.strip()
            
            # Jika baris kosong, berarti akhir dari sebuah kalimat
            if not line:
                if sentence_words:  # Hanya menambahkan jika ada kalimat
                    words.append(sentence_words)
                    pos_tags.append(sentence_pos_tags)
                sentence_words = []
                sentence_pos_tags = []
                continue
            
            # Memisahkan kolom
            columns = line.split('\t')
            if len(columns) >= 4:
                word = columns[1]  # Kata berada di kolom kedua
                pos_tag = columns[3]  # POS tag berada di kolom keempat
                sentence_words.append(word)
                sentence_pos_tags.append(pos_tag)
        
        # Menambahkan kalimat terakhir jika tidak ada baris kosong di akhir file
        if sentence_words:
            words.append(sentence_words)
            pos_tags.append(sentence_pos_tags)
    
    return words, pos_tags
```

```python
# Contoh penggunaan
conllu_file = 'data/raw/indonesian_pos_test.conllu'  # Ganti dengan nama file Anda
# conllu_file = 'data/raw/indonesian_pos_train.conllu'  
words, pos_tags = split_conllu_data(conllu_file)
```
